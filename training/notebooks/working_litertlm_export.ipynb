```python
MODEL = "V1rtucious/aegis-sft-e4b-merged-v4"
```

```python
from transformers import AutoTokenizer, AutoModelForImageTextToText

tok = AutoTokenizer.from_pretrained(MODEL)
cfg_model = AutoModelForImageTextToText.from_pretrained(
    MODEL,
    dtype="auto",
    device_map="auto",
)
print(cfg_model.config.model_type)
```

```python
del cfg_model
import gc, torch
gc.collect()
torch.cuda.empty_cache()
```

```python
from litert_torch.generative.export_hf.export import export

export(
    model=MODEL,
    output_dir="/content/gemma4_e4b_litertlm",
    task="image_text_to_text",
    prefill_lengths=[1024],
    cache_length=4096,
    quantization_recipe="dynamic_wi4_afp32",
    externalize_embedder=True,
    single_token_embedder=True,
    use_jinja_template=True,
    bundle_litert_lm=True,
    export_vision_encoder=False,
    keep_temporary_files=False,
)
```

```python
HF_REPO_ID = "V1rtucious/gemma4-e4b-toolcalling-litertlm-int4"
MODEL_PATH = "/content/gemma4_e4b_litertlm/model.litertlm"
```

```python
import os

assert os.path.exists(MODEL_PATH), f"Missing: {MODEL_PATH}"
print("Size GB:", os.path.getsize(MODEL_PATH) / 1e9)
```

```python
readme = """---
library_name: litert
base_model: google/gemma-4-E4B-it
tags:
- litert-lm
- gemma4
- text-generation
- tool-calling
- on-device
license: gemma
---

# Gemma 4 E4B Tool-Calling LiteRT-LM

Text-only LiteRT-LM export of a fine-tuned Gemma 4 E4B model for tool-calling.

## Files

- `model.litertlm`: LiteRT-LM bundle exported with vision disabled.

## Export Notes

- Task: text/tool-calling generation
- Vision export: disabled
- Quantization: dynamic_wi8_afp32
- Prefill length: 1024
- Cache length: 4096

This model is intended for apps that parse tool calls from generated text.
"""

with open("/content/gemma4_e4b_litertlm/README.md", "w") as f:
    f.write(readme)
```

```python
from huggingface_hub import HfApi, create_repo

api = HfApi()

create_repo(
    repo_id=HF_REPO_ID,
    repo_type="model",
    private=True,
    exist_ok=True,
)

api.upload_folder(
    folder_path="/content/gemma4_e4b_litertlm",
    repo_id=HF_REPO_ID,
    repo_type="model",
    commit_message="Upload LiteRT-LM text-only Gemma 4 E4B export",
)
```
